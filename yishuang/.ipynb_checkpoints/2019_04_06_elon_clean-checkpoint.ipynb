```python
elon = pd.read_csv("../../data/twitter/elon.csv", sep = ";",dtype = {'id':'object','reteets':'int','favorites':'int'})
```

```python
elon.describe()
```

```python
elon.head()
```

```python
index = elon[elon.id.isna()].index.tolist()
index
```

```python
elon[elon.id.isna()]
```

```python
pattern = re.compile('";(.*?);(.*?);(.*?);(.*?)"')
pattern2 = re.compile('"(.*?)"')
for i in range(len(index)):
    print(i)
    content = elon.iloc[index[i]]['text']
    result = re.findall(pattern, content)[0]
    new_content = content.replace('"'+re.findall(pattern2, content)[0]+'"','')
    elon.set_value(index[i],'permalink', elon.iloc[index[i]].geo)
    elon.set_value(index[i],'geo',result[0])
    elon.set_value(index[i],'mentions',result[1])
    elon.set_value(index[i],'hashtags',result[2])
    elon.set_value(index[i],'id',result[3])
    elon.set_value(index[i],'text',new_content)
    
```

```python
[elon.iloc[index[i]] for i in range(len(index))]
```

```python
pattern = re.compile('";(.*?);(.*?);(.*?);(.*?)"')
pattern2 = re.compile('"(.*?)"')
content = tesla.iloc[2632]['text']
result = re.findall(pattern, content)[0]
new_content = content.replace('"'+re.findall(pattern2, content)[0]+'"','')
tesla.set_value(2632,'permalink', tesla.iloc[2632].geo)
tesla.set_value(2632,'geo',result[0])
tesla.set_value(2632,'mentions',result[1])
tesla.set_value(2632,'hashtags',result[2])
tesla.set_value(2632,'id',result[3])
tesla.set_value(2632,'text',new_content)
```

```python
date = pd.to_datetime(elon.date).dt.date.unique()
```

```python
elon['adj_date'] = pd.to_datetime(elon.date).dt.date
```

```python
a = elon.groupby('adj_date')['text'].apply('\n'.join).reset_index()
```

```python
print(a.text[2])
```

```python
b = elon.groupby(by='adj_date').agg({'retweets':sum}).reset_index()
c = elon.groupby(by='adj_date').agg({'favorites':sum}).reset_index()
```

```python
elon.fillna("", inplace = True)
d = elon.groupby('adj_date')['mentions'].apply(' '.join).reset_index()
e = elon.groupby('adj_date')['hashtags'].apply(' '.join).reset_index()
f = elon.groupby('adj_date')['permalink'].apply(' '.join).reset_index()
```

```python
new = pd.merge(a, b, on = 'adj_date')
new = pd.merge(new, c, on = 'adj_date')
new = pd.merge(new, d, on = 'adj_date')
new = pd.merge(new, e, on = 'adj_date')
new_elon = pd.merge(new, f, on = 'adj_date')
new_elon
```

```python
new_elon.to_csv("../../data/twitter/elon_aggregate.csv", index = False)
```

```python
tesla['adj_date'] = pd.to_datetime(tesla.date).dt.date
```

```python
a = tesla.groupby('adj_date')['text'].apply('\n'.join).reset_index()
b = tesla.groupby(by='adj_date').agg({'retweets':sum}).reset_index()
c = tesla.groupby(by='adj_date').agg({'favorites':sum}).reset_index()
tesla.fillna("", inplace = True)
d = tesla.groupby('adj_date')['mentions'].apply(' '.join).reset_index()
e = tesla.groupby('adj_date')['hashtags'].apply(' '.join).reset_index()
f = tesla.groupby('adj_date')['permalink'].apply(' '.join).reset_index()
```

```python
new = pd.merge(a, b, on = 'adj_date')
new = pd.merge(new, c, on = 'adj_date')
new = pd.merge(new, d, on = 'adj_date')
new = pd.merge(new, e, on = 'adj_date')
new_tesla = pd.merge(new, f, on = 'adj_date')
new_tesla
```

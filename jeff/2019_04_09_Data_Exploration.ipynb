```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import datetime

```

```python
df_elon=pd.read_csv("../../data/twitter/Elon_0402.csv")
```

```python
df_elon.head()
```

```python
df_elon.info()
```

```python
df_elon['DateTime'] = pd.to_datetime(df_elon['date'])



df_elon['year']= df_elon['DateTime'].dt.year
df_elon['month']=df_elon['DateTime'].dt.month
df_elon['day']=df_elon['DateTime'].dt.day
df_elon.head()
```

```python
df_elon_short=df_elon[["retweets","favorites","text","DateTime","year","month","day"]].dropna()
df_elon_short.head()
```

```python
df_elon_short_sort=df_elon_short.sort_values(by=["retweets"],ascending=False).reset_index(drop=True)
df_elon_short_sort.head()
```

```python
df_elon_chairman=df_elon_short_sort[(df_elon_short_sort['DateTime']<"2018-12-01") & (df_elon_short_sort['DateTime'] >"2018-09-23")].reset_index(drop=True)
df_elon_chairman
```

```python
df_elon_events=df_elon_chairman[df_elon_chairman["text"].str.match("Tesla")]

df_elon_events
```

```python
df_tesla=pd.read_csv("../../data/twitter/tesla_clean.csv")
```

```python
df_tesla['DateTime'] = pd.to_datetime(df_tesla['date'])



df_tesla['year']= df_tesla['DateTime'].dt.year
df_tesla['month']=df_tesla['DateTime'].dt.month
df_tesla['day']=df_tesla['DateTime'].dt.day
df_tesla.head()
```

```python
df_tesla_chairman=df_tesla_short_sort[(df_tesla_short_sort['DateTime']<"2018-12-01") & (df_tesla_short_sort['DateTime'] >"2018-09-23")].reset_index(drop=True)
df_tesla_chairman
```

```python
df_tesla_events=df_tesla_chairman.iloc[:8,:]
df_tesla_events
```

```python
df_time=pd.merge(df_elon_events,df_tesla_events, on="year" and "month" and "day",how="inner")
df_time
```

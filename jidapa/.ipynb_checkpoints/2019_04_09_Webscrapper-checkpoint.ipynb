```python
import GetOldTweetspython.got3 as got
```

```python
import pandas as pd
topuser = pd.read_csv('./toptweet.csv')
topuser.head()
```

```python
import sys,getopt,datetime,codecs
outputFile = codecs.open('tweetFromTopUserElon.csv', "w+", "utf-8")
outputFile.write('username;date;retweets;favorites;text;geo;mentions;hashtags;id;permalink')

def receiveBuffer(tweets):
    for t in tweets:
        outputFile.write(('\n%s;%s;%d;%d;"%s";%s;%s;%s;"%s";%s' % (t.username, t.date.strftime("%Y-%m-%d %H:%M"), t.retweets, t.favorites, t.text, t.geo, t.mentions, t.hashtags, t.id, t.permalink)))
    outputFile.flush()
    print('More %d saved on file...\n' % len(tweets))

for _, t in topuser.iterrows():
    print('read from ',t.username)
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('elonmusk').setUsername(t.username).setSince("2010-01-01").setUntil("2019-04-02")
    tweet = got.manager.TweetManager.getTweets(tweetCriteria,receiveBuffer)
    if len(tweet) == 0:
        print('cannot find a related tweet')
        
outputFile.close() 
```

```python
import sys,getopt,datetime,codecs
outputFile = codecs.open('Elon.csv', "w+", "utf-8")
outputFile.write('username;date;retweets;favorites;text;geo;mentions;hashtags;id;permalink;isreply')

def receiveBuffer(tweets):
    for t in tweets:
        outputFile.write(('\n%s;%s;%d;%d;"%s";%s;%s;%s;"%s";%s;%s' % (t.username, t.date.strftime("%Y-%m-%d %H:%M"), t.retweets, t.favorites, t.text, t.geo, t.mentions, t.hashtags, t.id, t.permalink,t.isReply)))
    outputFile.flush()
    print('More %d saved on file...\n' % len(tweets))

tweetCriteria = got.manager.TweetCriteria().setUsername('elonmusk').setSince("2010-01-01").setUntil("2019-04-05")
tweet = got.manager.TweetManager.getTweets(tweetCriteria,receiveBuffer)
outputFile.close() 
```

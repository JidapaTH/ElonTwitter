### Get the comments

```python
# Due to RAM issue I split by month

import itertools
def getLastDateOfMonth(year: int, month: int):
    from datetime import date, timedelta
    """Given a year and a month returns an instance of the datetime.date class
    containing the last day of the corresponding month.

    Source: https://stackoverflow.com/a/43088/1445572
    """
    if month == 12:
        # last day of year is always 31st
        return date(year, month, 31)
    else:
        # go to the next month 1st day, and move back 1 day
        return date(year, month + 1, 1) - timedelta(days=1)
    
iterables = [ [2017],  [11] ]

date = [[t[0],t[1],getLastDateOfMonth(t[0],t[1]).day] for t in itertools.product(*iterables)]
```

```python
date = date[0:87] + [[2019,4,5]]
```

```python
import sys,getopt,datetime,codecs

for i in date:
    print("%d-%02d-%02d" % (i[0],i[1],i[2]))
    outputFile = codecs.open('../../data/Comment_to_Elon/to_Elon_%d-%02d.csv'% (i[0],i[1]), "w+", "utf-8")
    outputFile.write('username;date;retweets;favorites;text;geo;mentions;hashtags;id;permalink;isreply')

    def receiveBuffer(tweets):
        for t in tweets:
            outputFile.write(('\n%s;%s;%d;%d;"%s";%s;%s;%s;"%s";%s;%s' % (t.username, t.date.strftime("%Y-%m-%d %H:%M"), t.retweets, t.favorites, t.text, t.geo, t.mentions, t.hashtags, t.id, t.permalink,t.isReply)))
        outputFile.flush()

    tweetCriteria = got.manager.TweetCriteria().setCite('elonmusk').setSince("%d-%02d-01" % (i[0],i[1])).setUntil("%d-%02d-%02d" % (i[0],i[1],i[2]))
    tweet = got.manager.TweetManager.getTweets(tweetCriteria,receiveBuffer)
    outputFile.close() 
```
